```python
import pandas as pd

# 경로는 본인 PC 기준으로 수정
df = pd.read_csv("C:/Users/SAMSUNG/Desktop/AI/mypython/Python_Webscraping_Analysis/data/인구현황.csv", encoding='utf-8-sig')

# '전국' 제외
region_df = df[df['행정기관'] != '전국'].copy()

# 총인구수 컬럼이 문자열이면 쉼표 제거 후 숫자형으로 변환
if region_df['총인구수'].dtype == 'object':
    region_df['총인구수'] = region_df['총인구수'].str.replace(',', '').astype(int)

# 총인구수 상위 5개 추출
top5 = region_df.sort_values(by='총인구수', ascending=False).head(5)

# 결과 출력
print("지역별 총 인구수 TOP 5:")
print(top5[['행정기관', '총인구수']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 지역 개수만큼 색상 만들기 (예: 17개 지역 → 17가지 색상)
num_regions = region_df.shape[0]
colors = sns.color_palette("hls", num_regions)

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(data=region_df, x='행정기관', y='총인구수', palette=colors)

# 제목 및 레이블
plt.title('지역별 총인구수 (명확한 색상)', fontsize=16)
plt.xlabel('지역')
plt.ylabel('총인구수')
plt.xticks(rotation=45)
plt.tight_layout()

# 출력
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 정렬 (옵션: 비율이 큰 순서로 보기 좋게)
sorted_df = region_df.sort_values(by='남여 비율', ascending=False)

# 색상 팔레트
colors = sns.color_palette("coolwarm", len(sorted_df))

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(data=sorted_df, x='행정기관', y='남여 비율', palette=colors)

# 기준선 추가 (비율 1.0 = 남녀 동일)
plt.axhline(y=1.0, color='red', linestyle='--', linewidth=2, label='남녀 동일 기준 (1.0)')

# 그래프 제목 및 축 설정
plt.title('행정기관별 남여 비율 분석', fontsize=16)
plt.xlabel('지역')
plt.ylabel('남여 비율 (남성 / 여성)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import pandas as pd

# 새로운 컬럼 '남초여초' 생성
def classify_gender_ratio(ratio):
    if ratio > 1.0:
        return '남초'
    elif ratio < 1.0:
        return '여초'
    else:
        return '동일'

# 필요한 컬럼 선택 및 분류 적용
gender_category_df = region_df[['행정기관', '남여 비율']].copy()
gender_category_df['남초여초'] = gender_category_df['남여 비율'].apply(classify_gender_ratio)

# 결과 출력
print("✅ 남자초과 / 여자초과 분류표")
print(gender_category_df)
```

```python
import pandas as pd

# 전국 평균 세대당 인구 계산
average_household_size = region_df['세대당 인구'].mean()

# 평균보다 높은 지역 필터링
above_avg_df = region_df[region_df['세대당 인구'] > average_household_size][['행정기관', '세대당 인구']].sort_values(by='세대당 인구', ascending=False)

# 전국 평균 출력
print(f"✅ 전국 평균 세대당 인구수: {average_household_size:.2f}\n")

# 결과 표 출력
print("🏠 세대당 인구 평균보다 높은 지역:")
print(above_avg_df)
```

```python
# 남자/여자 인구 컬럼 숫자형으로 변환
for col in ['남자 인구수', '여자 인구수']:
    if region_df[col].dtype == 'object':
        region_df[col] = region_df[col].str.replace(',', '').astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 재구성 (성별 비교용 melt)
gender_compare_df = region_df[['행정기관', '남자 인구수', '여자 인구수']].melt(
    id_vars='행정기관',
    var_name='성별',
    value_name='인구수'
)

# 그래프 그리기
plt.figure(figsize=(14, 6))
sns.barplot(data=gender_compare_df, x='행정기관', y='인구수', hue='성별', palette=['skyblue', 'lightcoral'])

# 제목 및 축 설정
plt.title('지역별 남자 인구 vs 여자 인구 비교', fontsize=16)
plt.xlabel('지역')
plt.ylabel('인구수')
plt.xticks(rotation=45)
plt.legend(title='성별')
plt.tight_layout()
plt.show()
```
